Question 1 : What is Simple Linear Regression (SLR)? Explain its purpose

Ans:- Simple Linear Regression (SLR) is a statistical and machine-learning technique used to model the relationship between one independent variable (X) and one dependent variable (Y) by fitting a straight line to the observed data.

**Purpose of Simple Linear Regression**

1. Understand Relationships

It helps in understanding how changes in one variable (X) affect another variable (Y).

2. Prediction

Once the model is trained, it can be used to predict the value of Y for a given value of X.

3. Quantify Impact

Question 2: What are the key assumptions of Simple Linear Regression?
Ans:- The key assumptions of Simple Linear Regression (SLR) ensure that the model produces reliable and unbiased results. These assumptions are:

1. Linearity

There is a linear relationship between the independent variable (X) and the dependent variable (Y).

This means Y changes at a constant rate with respect to X.

Can be checked using a scatter plot.

2. Independence of Errors

The residuals (errors) are independent of each other.

No correlation between consecutive errors.

Important especially for time-series data.

3. Homoscedasticity

The residuals have constant variance across all levels of X.

Errors should be spread evenly around the regression line.

Unequal spread indicates heteroscedasticity.

4.  Normality of Errors

The residuals are normally distributed.

This is mainly required for hypothesis testing and confidence intervals.

Checked using histograms or Q‚ÄìQ plots.

5. No Multicollinearity (Automatically satisfied in SLR)

Since SLR has only one independent variable, multicollinearity is not an issue.

This assumption becomes important in multiple linear regression.

6. Zero Mean of Errors

The average of the error terms is zero, meaning the model is unbiased.

Question 3: Write the mathematical equation for a simple linear regression model and explain each term.

Ans:-
Y = Œ≤0 + Œ≤1X + Œµ

Here's what each term means:
Y: Dependent variable (the outcome we're trying to predict)
X: Independent variable (the predictor)
Œ≤0: Intercept or constant term (where the line crosses the Y-axis)
Œ≤1: Slope coefficient (change in Y for a one-unit change in X)
Œµ: Error term (the random variation in Y not explained by X)

Question 5: What is the method of least squares in linear regression?

Ans:-
The method of least squares is a way to find the best-fitting line in linear regression. It works by minimizing the sum of the squared differences between actual data points and predicted values.
In simple terms, it:
Calculates the distance between each data point and the predicted line
Squares each of these distances
Adds them all up
Adjusts the line to make this sum as small as possible
This gives us the line that best fits the data!

Question 6: What is Logistic Regression? How does it differ from Linear Regression?

Ans:-Logistic Regression is a statistical model used for classification problems, where the outcome (Y) is categorical (e.g., 0/1, yes/no, spam/not spam). It's used to predict the probability of an event occurring based on input variables (X).
Key differences from Linear Regression:
- Outcome variable: Logistic Regression is for categorical outcomes, while Linear Regression is for continuous outcomes.
- Prediction: Logistic Regression predicts probabilities, while Linear Regression predicts a continuous value.
- Equation: Logistic Regression uses a logistic function (S-shaped curve) to model probabilities, whereas Linear Regression uses a straight line.
Example: Predicting whether a customer will buy a product (yes/no) based on their age and income.

Question 7: Name and briefly describe three common evaluation metrics for regression models?

Ans:-
1. Mean Absolute Error (MAE): Average difference between predicted and actual values.
2. Mean Squared Error (MSE): Average of squared differences between predicted and actual values.
3. R-squared (R¬≤): Measures how well the model explains the variance in the data (0-1, higher is better).


Question 8: What is the purpose of the R-squared metric in regression analysis?

Ans:- R-squared (R¬≤) measures how well the model fits the data. It tells you the proportion of variance in the dependent variable (Y) that's explained by the independent variable(s) (X).
In simple terms, it answers: "How much of the change in Y is explained by X?"
R¬≤ values range from 0 to 1:
- 0: Model explains none of the variance
- 1: Model explains all the variance
- Higher R¬≤ generally indicates a better fit, but be cautious of overfitting!

```python
# Question 9: Write Python code to fit a simple linear regression model using scikit-learn and print the slope and intercept.

import numpy as np
from sklearn.linear_model import LinearRegression

# Sample data (replace with your own)
X = np.array([[1], [2], [3], [4], [5]])  # independent variable
y = np.array([2, 3, 5, 7, 11])  # dependent variable

# Create and fit the model
model = LinearRegression()
model.fit(X, y)

# Print slope and intercept
print("Slope (Œ≤1):", model.coef_[0])
print("Intercept (Œ≤0):", model.intercept_)
```

Question 10: How do you interpret the coefficients in a simple linear regression model?

Ans:- Interpretation of Coefficients in a Simple Linear Regression Model

In a simple linear regression model, the coefficients have straightforward interpretations.
- The slope coefficient (Œ≤1) represents the change in the dependent variable (Y) for a one-unit change in the independent variable (X), while holding all other variables constant (which isn't really applicable here since it's simple regression).
- The intercept coefficient (Œ≤0) represents the expected value of Y when X is 0.

Mathematically, you can express the model as Y = Œ≤0 + Œ≤1X + Œµ, where Œµ is the error term.
